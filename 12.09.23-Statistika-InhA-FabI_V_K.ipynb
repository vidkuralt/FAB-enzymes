# Normality testing-InhA_a

```python
# D'Agostino and Pearson's Test
# normality test
statistics=[]
p_value=[]
list
alpha=0.05
for i in continuous_features:
    stat, p = normaltest(df_InhA_a[[i]], axis=0, nan_policy='propagate') 
    statistics.append(stat)
    p_value.append(p)
      # interpret results
    if p > alpha:
        print(i, stat, np.round(p,4), 'Normal distribution')
    else:
           print(i, stat, np.round(p,4), 'Non-Normal distribution') 
```

```python
def Holm(v):
    vout = []
    v.sort()
    for i in range(0,len(v),1):
        if v[i]<=(0.05/(len(v)-i)):
            vout.append(v[i])
        else:
            break
    return vout
```

```python
cf=np.array(continuous_features)
cf =pd.DataFrame(cf)
```

```python
p_value_norm = np.round(p_value,3)
p_value_norm=pd.DataFrame(p_value_norm)
p_value_norm.columns=['p_value']
```

```python
print(p_value_norm)
```

```python
InhA_a=pd.concat([cf, p_value_norm], axis=1)
```

# Kruskal-Wallis H-test

The Kruskal-Wallis H-test tests the null hypothesis that the population median of all of the groups are equal. It is a non-parametric version of ANOVA. The test works on 2 or more independent samples, which may have different sizes. Note that rejecting the null hypothesis does not indicate which of the groups differs. Post hoc comparisons between groups are required to determine which groups are different.

# Vsi encimi

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import mannwhitneyu
from statsmodels.stats.multitest import multipletests

# List of enzymes
enzymes = ['FabI_a', 'InhA_a', 'FabV_a', 'FabK_a']

# Dictionary to store p-value matrices for each feature
p_values_matrices = {}

# Dictionary to store significant matrices for each feature
significant_matrices = {}

# Loop through all features
for feature in features2:
    # Create an empty DataFrame with the enzymes as indices and columns
    p_values_matrix = pd.DataFrame(index=enzymes, columns=enzymes)

    # Loop through all pairs of enzymes
    for i in range(len(enzymes)):
        for j in range(i + 1, len(enzymes)):
            enzyme1 = enzymes[i]
            enzyme2 = enzymes[j]

            # Skip if enzyme1 or enzyme2 is equal to the current enzyme
            if enzyme1 == enzyme2:
                continue

            # Extract the values for the current feature and the two enzymes
            values1 = df1[df1['Activity3'] == enzyme1][feature]
            values2 = df1[df1['Activity3'] == enzyme2][feature]

            
            # Calculate the Mann-Whitney U test p-value
            _, p_value = mannwhitneyu(values1, values2)

            # Fill in the corresponding cell of the DataFrame
            p_values_matrix.loc[enzyme1, enzyme2] = p_value
            p_values_matrix.loc[enzyme2, enzyme1] = p_value

    # Convert NaN values to 1
    p_values_matrix = p_values_matrix.fillna(1)

    # Apply Holm-Bonferroni correction
    p_values = p_values_matrix.values.flatten()
    corrected_p_values = multipletests(p_values, method='holm')[1]
    corrected_p_values_matrix = pd.DataFrame(corrected_p_values.reshape(p_values_matrix.shape), index=enzymes, columns=enzymes)

    # Create a significant matrix based on corrected p-values
    significant_matrix = corrected_p_values_matrix <= 0.05
    significant_matrix = significant_matrix.replace({True: 'S', False: 'NS'})

    # Convert NaN values to NS
    significant_matrix = significant_matrix.fillna('NS')

    # Store the matrices in the dictionaries
    p_values_matrices[feature] = p_values_matrix
    significant_matrices[feature] = significant_matrix

    # Save the matrices as CSV files
    os.chdir(PATH)
    if not os.path.exists("MannWhitney_vsi"):
                    os.makedirs("MannWhitney_vsi")

    p_values_matrix.to_csv(os.path.join(PATH, 'MannWhitney_vsi/p_values_{}.csv'.format(feature)))
    significant_matrix.to_csv(os.path.join(PATH, 'MannWhitney_vsi/significant_{}.csv'.format(feature)))
    p_values_matrix.to_excel(os.path.join(PATH, 'MannWhitney_vsi/p_values_{}.xlsx'.format(feature)))
    significant_matrix.to_excel(os.path.join(PATH, 'MannWhitney_vsi/significant_{}.xlsx'.format(feature)))

    # Create a heatmap and save it to a PNG file
    sns.set(font_scale=1.2)
    sns.heatmap(p_values_matrix, annot=True, cmap='Blues', fmt='.3f', square=True, cbar=False)
    plt.title(feature)

    plt.savefig(os.path.join(PATH, 'MannWhitney_vsi/Heatmap_'+str(feature)+'.png'))
    plt.show()
for feature, p_values_matrix in p_values_matrices.items():
    print("Feature:", feature)
    print(p_values_matrix)
```
